# Example on INTC

```python
#make sure date is correct data type
intc_data = pd.read_csv("INTC.csv")
intc_data
```

```python
#make sure index is datetime
intc_data['Date'] = pd.to_datetime(intc_data['Date'])
intc_data.set_index('Date', inplace=True)
intc_data.ffill(inplace=True)
intc_data
```

```python
def compute_rsi(series, period=14):
    delta = series.diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def compute_macd(series, short_period=12, long_period=26, signal_period=9):
    short_ema = series.ewm(span=short_period, adjust=False).mean()
    long_ema = series.ewm(span=long_period, adjust=False).mean()
    macd = short_ema - long_ema
    signal_line = macd.ewm(span=signal_period, adjust=False).mean()
    return macd, signal_line

def compute_bollinger_bands(series, window=20, num_std_dev=2):
    middle_band = series.rolling(window=window).mean()
    std_dev = series.rolling(window=window).std()
    upper_band = middle_band + (num_std_dev * std_dev)
    lower_band = middle_band - (num_std_dev * std_dev)
    return middle_band, upper_band, lower_band

def compute_obv(close, volume):
    obv = [0]
    for i in range(1, len(close)):
        if close[i] > close[i - 1]:
            obv.append(obv[-1] + volume[i])
        elif close[i] < close[i - 1]:
            obv.append(obv[-1] - volume[i])
        else:
            obv.append(obv[-1])
    return pd.Series(obv, index=close.index)

def compute_vwap(close, volume):
    cumulative_volume = volume.cumsum()
    cumulative_price_volume = (close * volume).cumsum()
    vwap = cumulative_price_volume / cumulative_volume
    return vwap

def compute_ad_line(high, low, close, volume):
    money_flow_multiplier = ((close - low) - (high - close)) / (high - low)
    money_flow_volume = money_flow_multiplier * volume
    ad_line = money_flow_volume.cumsum()
    return ad_line

def compute_cmf(high, low, close, volume, period=20):
    money_flow_multiplier = ((close - low) - (high - close)) / (high - low)
    money_flow_volume = money_flow_multiplier * volume
    cmf = money_flow_volume.rolling(window=period).sum() / volume.rolling(window=period).sum()
    return cmf

def compute_volume_oscillator(volume, short_period=5, long_period=20):
    short_volume_ma = volume.rolling(window=short_period).mean()
    long_volume_ma = volume.rolling(window=long_period).mean()
    volume_oscillator = (short_volume_ma - long_volume_ma) / long_volume_ma * 100
    return volume_oscillator

#Feature engineering : calculate technical indicators
intc_data['20D_MA'] = intc_data['Close'].rolling(window=20).mean() #20 day moving average
intc_data['50D_MA'] = intc_data['Close'].rolling(window=50).mean() #50 day moving average
intc_data['Daily_Return'] = intc_data['Close'].pct_change() #Daily Return
intc_data['EMA_20'] = intc_data['Close'].ewm(span=20, adjust=False).mean()
intc_data['RSI'] = compute_rsi(intc_data['Close'])
intc_data['MACD'], intc_data['MACD_Signal'] = compute_macd(intc_data['Close'])
intc_data['BB_Middle'], intc_data['BB_Upper'], intc_data['BB_Lower'] = compute_bollinger_bands(intc_data['Close'])
intc_data['OBV'] = compute_obv(intc_data['Close'], intc_data['Volume'])
intc_data['VWAP'] = compute_vwap(intc_data['Close'], intc_data['Volume'])
intc_data['AD_Line'] = compute_ad_line(intc_data['High'], intc_data['Low'], intc_data['Close'], intc_data['Volume'])
intc_data['CMF'] = compute_cmf(intc_data['High'], intc_data['Low'], intc_data['Close'], intc_data['Volume'])
intc_data['Volume_Oscillator'] = compute_volume_oscillator(intc_data['Volume'])
intc_data['Volume_MA_20'] = intc_data['Volume'].rolling(window=20).mean()
intc_data['Close_Lag1'] = intc_data['Close'].shift(1)
intc_data['Close_Lag2'] = intc_data['Close'].shift(2)
intc_data['Close_Lag3'] = intc_data['Close'].shift(3)
intc_data
```

```python
from sklearn.preprocessing import MinMaxScaler

#Normalize the input features
feature_columns = [
    'Close', 'High', 'Low', 'Open', 'Volume', '20D_MA', '50D_MA', 'Daily_Return', 'Close_Lag1',
    'EMA_20', 'RSI', 'MACD', 'MACD_Signal', 'BB_Middle', 'BB_Upper', 'BB_Lower',
    'OBV', 'VWAP', 'AD_Line', 'CMF', 'Volume_Oscillator', 'Volume_MA_20', 'Close_Lag2', 'Close_Lag3'
]

scaler = MinMaxScaler()
intc_data[feature_columns] = scaler.fit_transform(intc_data[feature_columns])
y_scaler = MinMaxScaler()
intc_data.dropna(inplace=True)
```

```python
import numpy as np

#Create Sequences
sequence_len = 90
X = []
y = []

for i in range(len(intc_data) - sequence_len):
    X.append(intc_data[feature_columns].iloc[i:i+sequence_len].values)
    y.append(intc_data['Adj Close'].iloc[i+sequence_len])

X = np.array(X)
y = np.array(y)

y_scaler = MinMaxScaler()
y = y_scaler.fit_transform(y.reshape(-1, 1)).flatten()

X.shape, y.shape
```

# Build an LSTM for INTC

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42, shuffle=False)
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,  # Reduce patience to stop earlier
    restore_best_weights=True
)

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 0.0005 * 0.95 ** epoch)
```

```python
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Bidirectional

model = Sequential()
model.add(Bidirectional(LSTM(units=128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]))))
model.add(Dropout(0.2))
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(0.2)) 
model.add(LSTM(units=64, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))



model.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error')
```

```python
#Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X_test)
y_test = y_scaler.inverse_transform(y_test.reshape(-1,1)).flatten()
y_pred = y_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()

plt.figure(figsize=(14, 5))
plt.plot(y_test, label='Actual Values', color='blue')
plt.plot(y_pred, label='Predicted Values', color='orange')
plt.title('Actual vs. Predicted Values on Test Data')
plt.xlabel('Sample Index')
plt.ylabel('Adj Close Price')
plt.legend()
plt.show()
```
