# Feature Engineering
- Focus on returns rather than numbers

```python
intc_data['Daily_Return'] = intc_data['Close'].pct_change()
intc_data
```

Lagged Daily Returns:

Use past returns as input features to help the model identify patterns.
Features: Daily_Return, Return_Lag1, Return_Lag2, Return_Lag3, etc.

```python
for i in range(1, 5):
    intc_data[f'Return_Lag{i}'] = intc_data['Daily_Return'].shift(i)

intc_data
```

#### Momentum Indicators (Rate of Change):
- These are already expressed as a percentage change over time.
- Examples:
= Rate of Change (ROC):

`intc_data['ROC_5'] = intc_data['Close'].pct_change(periods=5)`
`intc_data['ROC_10'] = intc_data['Close'].pct_change(periods=10)`

#### Moving Averages (in Returns):

- Calculate moving averages on the daily returns.
- Examples:

`intc_data['MA_Return_5'] = intc_data['Daily_Return'].rolling(window=5).mean()`
`intc_data['MA_Return_10'] = intc_data['Daily_Return'].rolling(window=10).mean()`

#### Volatility (Standard Deviation of Returns):

- This can indicate how much the stock's return varies over time.
- Examples:

`intc_data['Volatility_5'] = intc_data['Daily_Return'].rolling(window=5).std()`
`intc_data['Volatility_10'] = intc_data['Daily_Return'].rolling(window=10).std()`

#### Cumulative Return:
- The cumulative return is the compounded return over a period, but expressed in a relative change.
- Example:

`intc_data['Cumulative_Return'] = (1 + intc_data['Daily_Return']).cumprod() - 1`

#### Relative Strength Index (RSI):
- RSI is bounded between 0 and 100 and represents relative changes in price, so it aligns with the scale of returns.
- Code:

`intc_data['RSI'] = compute_rsi(intc_data['Close'])`

#### On-Balance Volume (OBV):

- While OBV is cumulative, you can convert it to daily changes.
- Example:

`intc_data['OBV'] = compute_obv(intc_data['Close'], intc_data['Volume'])`
`intc_data['OBV_Change'] = intc_data['OBV'].pct_change()  # Convert OBV to daily changes`

#### MACD (in Returns):
- The MACD and signal line are difference-based indicators and can be used directly.
- Code:

`intc_data['MACD'], intc_data['MACD_Signal'] = compute_macd(intc_data['Close'])`

```python
import numpy as np

def compute_rsi(series, period=14):
    delta = series.diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def compute_obv(close, volume):
    obv = [0]
    for i in range(1, len(close)):
        if close[i] > close[i - 1]:
            obv.append(obv[-1] + volume[i])
        elif close[i] < close[i - 1]:
            obv.append(obv[-1] - volume[i])
        else:
            obv.append(obv[-1])
    return pd.Series(obv, index=close.index)

def compute_macd(series, short_period=12, long_period=26, signal_period=9):
    short_ema = series.ewm(span=short_period, adjust=False).mean()
    long_ema = series.ewm(span=long_period, adjust=False).mean()
    macd = short_ema - long_ema
    signal_line = macd.ewm(span=signal_period, adjust=False).mean()
    return macd, signal_line

# Calculate additional return-based features
intc_data['ROC_5'] = intc_data['Close'].pct_change(periods=5)
intc_data['MA_Return_5'] = intc_data['Daily_Return'].rolling(window=5).mean()
intc_data['Volatility_5'] = intc_data['Daily_Return'].rolling(window=5).std()
intc_data['Volatility_10'] = intc_data['Daily_Return'].rolling(window=10).std().fillna(0)
intc_data['RSI'] = compute_rsi(intc_data['Close'])
intc_data['OBV'] = compute_obv(intc_data['Close'], intc_data['Volume']).pct_change()
intc_data['MACD'], intc_data['MACD_Signal'] = compute_macd(intc_data['Close'])

# Scale RSI to be between 0 and 1
intc_data['RSI'] = intc_data['RSI'] / 100.0


intc_data.replace([np.inf, -np.inf], np.nan, inplace=True)
intc_data.dropna(inplace=True)


intc_data
```

# Build a Model for INTC

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42, shuffle=False)
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 10],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'reg_alpha': [0, 0.01, 0.1, 1],
    'reg_lambda': [0.1, 1, 10]
}

# Initialize the model
xgb_model = XGBRegressor(random_state=42)

# Set up RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=50,  # Number of random combinations to try
    scoring='neg_mean_squared_error',
    cv=3,  # Cross-validation folds
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit the model
random_search.fit(X_train, y_train)

# Best parameters and model
best_xgb_model = random_search.best_estimator_
print(f"Best Parameters: {random_search.best_params_}")
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Initialize the model
rf = RandomForestRegressor(
    random_state=42)

# Set up RandomizedSearchCV
rf_random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_grid,
    n_iter=50,  # Number of random combinations to try
    scoring='neg_mean_squared_error',
    cv=3,  # Cross-validation folds
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit the model
rf_random_search.fit(X_train, y_train)

# Print the best parameters
print(f"Best Parameters: {rf_random_search.best_params_}")

# Get the best model
best_rf_model = rf_random_search.best_estimator_
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.metrics import root_mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
import numpy as np

# Create a function to find the optimal polynomial degree
def find_best_polynomial_degree(X_train, y_train, X_test, y_test, max_degree=64):
    best_degree = 1
    best_score = float('inf')

    for degree in range(1, max_degree + 1):
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse = root_mean_squared_error(y_test, y_pred)

        print(f'Degree {degree} RMSE: {mse:.4f}')
        if mse < best_score:
            best_score = mse
            best_degree = degree

    return best_degree, best_score

# Generate meta-features from the base model predictions
meta_X_train = np.column_stack([best_xgb_model.predict(X_train), best_rf_model.predict(X_train)])
meta_X_test = np.column_stack([best_xgb_model.predict(X_test), best_rf_model.predict(X_test)])

# Find the optimal degree for the polynomial regression
best_degree, best_score = find_best_polynomial_degree(meta_X_train, y_train, meta_X_test, y_test)

# Train the final polynomial regression model with the best degree
meta_model = make_pipeline(PolynomialFeatures(best_degree), LinearRegression())
meta_model.fit(meta_X_train, y_train)

# Create the stacking regressor with the tuned RF model
stacking_regressor = StackingRegressor(
    estimators=[
        ('xgb', best_xgb_model), 
        ('rf', best_rf_model)],
    final_estimator=meta_model
)

# Create the stacking regressor with the tuned RF model
stacking_regressor = StackingRegressor(
    estimators=[
        ('xgb', best_xgb_model), 
        ('rf', best_rf_model)],
    final_estimator=meta_model
)

# Fit the stacking model
stacking_regressor.fit(X_train, y_train)
```

```python
# Evaluate and make predictions
y_pred = stacking_regressor.predict(X_test)

# Plot and print the results
from sklearn.metrics import root_mean_squared_error

mse = root_mean_squared_error(y_test, y_pred)
print(f'Stacking Model Root Mean Squared Error: {mse}')

# Define a threshold for accuracy (e.g., within 0.01 of the actual value)
threshold = 0.01
accuracy_within_threshold = (abs(y_test - y_pred) <= threshold).mean() * 100

print(f"Percentage of predictions within ±{threshold}: {accuracy_within_threshold:.2f}%")

threshold = 0.05
accuracy_within_threshold = (abs(y_test - y_pred) <= threshold).mean() * 100
print(f"Percentage of predictions within ±{threshold}: {accuracy_within_threshold:.2f}%")

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(y_test.values, label='Actual Return', color='blue')
plt.plot(y_pred, label='Stacking Predicted Return', color='orange')
plt.title('Actual vs. Stacking Predicted Returns')
plt.xlabel('Sample Index')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```
